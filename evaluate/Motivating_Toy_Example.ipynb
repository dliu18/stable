```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results_df = pd.read_csv("link-prediction-toy.csv")
```

```python
plt.rc('font', size=16)          # controls default text sizes
plt.rc('axes', titlesize=16)     # fontsize of the axes title
plt.rc('axes', labelsize=16)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=14)    # fontsize of the tick labels
plt.rc('ytick', labelsize=14)    # fontsize of the tick labels
```

```python
for graphName in ["Facebook"]:
    fig, ax_roc = plt.subplots(figsize=(8,5))
    ax_size = ax_roc.twinx()
    
    sub_df = results_df[results_df["Graph"] == graphName]
    
    roc_handle, = ax_roc.plot(sub_df["k"], sub_df["ROC"], label="ROC", color="red")
    size_handle, = ax_size.plot(sub_df["k"], sub_df["Size"], label="Size")

    #results_df[results_df["Graph"] == graphName].plot(x="k", y="Size", ax=ax_size, label="Size")

    #ax.set(yscale="log")
    ax_roc.set_xlabel("K-core")
    ax_roc.set_ylabel("Link Prediction AUC-ROC")
    
    ax_size.set_ylabel("Fraction of Graph Remaining")
    #ax_size.set(yscale="log")
    
    ax_roc.legend(handles=[roc_handle, size_handle])
    ax_roc.set_title("Laplacian-Eigenmap Link Prediction Performance \nfor each K-core of a {} network".format(graphName))
```
