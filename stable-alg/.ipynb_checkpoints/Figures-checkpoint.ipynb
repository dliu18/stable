```python
import pickle
import pandas as pd
```

```python
with open("error_reductions.pickle", "rb") as pickleFile:
    error_reductions = pickle.load(pickleFile)
```

```python
error_reductions_pd = pd.DataFrame(error_reductions)
error_reductions_pd
```

```python
idx_to_graph = [3, 6, 9, 12, 15, 18, 21, 24, 27, 30]
error_reductions_subset = error_reductions_pd.iloc[idx_to_graph]
error_reductions_subset["Graph"] = [title.split("_")[0] for title in error_reductions_subset["Graph"]]
```

```python

error_reductions_subset[error_reductions_subset["Alg"] == "LE"].plot.bar(x="Graph", y="Error Reduction")
```

```python

error_reductions_subset[error_reductions_subset["Alg"] == "LINE"].plot.bar(x="Graph", y="Error Reduction")
```
