```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open("error_reductions_d32.pickle", "rb") as pickleFile:
    error_reductions = pickle.load(pickleFile)
```

```python
error_reductions_pd = pd.DataFrame(error_reductions)
error_reductions_pd
```

```python
idx_to_graph = [3, 6, 9, 12, 15, 18, 21, 24, 27, 30]
error_reductions_subset = error_reductions_pd.iloc[idx_to_graph]
error_reductions_subset["Graph"] = [title.split("_")[0] for title in error_reductions_subset["Graph"]]
```

```python
plt.rc('font', size=18)          # controls default text sizes
plt.rc('axes', titlesize=18)     # fontsize of the axes title
#plt.rc('axes', titleweight="bold")     # fontweight of the axes title
plt.rc('axes', labelsize=14)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=14)    # fontsize of the tick labels
plt.rc('ytick', labelsize=14)    # fontsize of the tick labels

fig, axs = plt.subplots(ncols=2, figsize=(14,5))
error_reductions_subset[error_reductions_subset["Alg"] == "LE"].plot.bar(x="Graph", 
                                                                         y="Error Reduction", 
                                                                         legend=False, 
                                                                         ax=axs[0])
error_reductions_subset[error_reductions_subset["Alg"] == "LINE"].plot.bar(x="Graph", 
                                                                           y="Error Reduction", 
                                                                           legend=False, 
                                                                           ax=axs[1])

axs[0].set_ylabel("Optimized / Original Stability Error")
axs[1].set_ylabel("Optimized / Original Stability Error")

axs[0].set_title("Laplacian Eigenmap")
axs[1].set_title("LINE")

for ax in axs:
    ax.set_xticklabels(ax.get_xticklabels(), rotation = 30)
    ax.set_xticklabels(ax.get_xticklabels(), rotation = 30)
```
